```python
from scipy import signal
import random


def f_high(y,sr):
    b,a = signal.butter(10, 2000/(sr/2), btype='highpass')
    yf = signal.lfilter(b,a,y)
    return yf
```

```python
yf1 = f_high(y1, sr)
```

```python

librosa.display.waveplot(yf1,sr=sr, x_axis='time');
```
